# Create a Class / 创建一个类
REVIEW: Building a neural network is made simple in PyTorch. / 复习：在 PyTorch 中，构建神经网络变得很简单。

This is because of `nn.Module`, which we inherit when creating our class to simplify building, managing, and organizing the model. / 这是因为我们在创建类时会继承 `nn.Module`，从而简化模型的构建、管理和组织。

This acts as the blueprint for our model. / 它相当于模型的蓝图。

### Structure of our Class / 类的结构

`__init__()`: This is where we define the layers of our network. / `__init__()`：在这里定义网络各层。

`forward()`: This is where we define how data flows through layers to produce predictions. / `forward()`：在这里定义数据如何经过各层并产生预测。


````python
# Import the nn module / 导入 nn 模块
import torch.nn as nn 
````

````python
# Create a simple class / 创建一个简单的类
class SimpleNeuralNetwork(nn.Module):
    def __init__(self):
        super(SimpleNeuralNetwork, self).__init__() # initialize superclass for automatic parameters / 初始化父类以自动管理参数
        
        # Define the layers: an input layer, a hidden layer, and an output layer
        # 定义层：输入层、隐藏层和输出层
        self.input_layer = nn.Linear(10, 20)  # Input size of 10, output size of 20 / 输入10维，输出20维
        self.hidden_layer = nn.Linear(20, 15) # Hidden layer with input size of 20, output size of 15 / 隐藏层输入20维，输出15维
        self.output_layer = nn.Linear(15, 1)  # Output layer with input size of 15, output size of 1 / 输出层输入15维，输出1维
        
        # Define the activation function (introduces non-linearity into the model)
        # 定义激活函数（为模型引入非线性）
        self.activation = nn.ReLU()

    # Define the forward pass / 定义前向传播
    def forward(self, x):
        x = self.activation(self.input_layer(x))  # Pass data through the input layer / 数据通过输入层
        x = self.activation(self.hidden_layer(x)) # Pass data through the hidden layer / 数据通过隐藏层
        x = self.output_layer(x)                  # Pass data through the output layer (no activation here) / 数据通过输出层（此处不加激活）
        return x
````

````python
# Demonstrate our model with an example / 使用示例演示模型
import torch

# Create a tensor with shape (5, 10) - batch of 5 samples, each with 10 features
# 创建形状为 (5, 10) 的张量：5个样本的批次，每个样本10个特征
example_tensor = torch.randn(5, 10)
print(example_tensor.size())
````

````python
# Create the input layer as it is in our Class / 按照类中的定义创建输入层
input_layer = nn.Linear(10, 20)

# Run the example through our input layer / 让示例数据通过输入层
input_linear_example = input_layer(example_tensor)
print(input_linear_example.size())
````

````python
# Do the same but with the hidden layer / 对隐藏层执行同样操作
hidden_layer = nn.Linear(20, 15)

# Run the input_linear_example through hidden layer / 将输入层输出传入隐藏层
hidden_linear_example = hidden_layer(input_linear_example)
print(hidden_linear_example.size())
````

````python
# Same for the output layer / 对输出层执行同样操作
output_layer = nn.Linear(15, 1)

# Run hidden_linear_example through output layer / 将隐藏层输出传入输出层
ouput_linear_example = output_layer(hidden_linear_example)
print(ouput_linear_example.size())
````

# 理解 ReLU 激活函数 / Understanding ReLU Activation

## ReLU 是什么？/ What is ReLU?

**ReLU (Rectified Linear Unit)** 是一个激活函数，公式很简单：

$$\text{ReLU}(x) = \max(0, x)$$

- **如果输入 > 0**：直接输出该值
- **如果输入 ≤ 0**：输出为 0

## 为什么需要激活函数？/ Why Activation Functions?

没有激活函数的神经网络只是多个线性变换的叠加，最终还是线性的。激活函数引入**非线性**，让模型能够学习复杂的模式。

## ReLU 在何时起作用？/ When ReLU is Applied?

在你的网络中：

```
输入 (5, 10)
    ↓
[线性层] → [ReLU激活] ← 这里起作用
    ↓
[线性层] → [ReLU激活] ← 这里起作用  
    ↓
[线性层] (无激活)
    ↓
输出 (5, 1)
```

**关键点：**
- ReLU 在**线性层之间**应用
- 它改变数据，然后传给下一層
- 最后输出层通常**不加激活**


````python
import torch
import torch.nn as nn

print("=== ReLU 激活函数的具体效果 ===\n")

# 创建一个包含正数和负数的示例张量
test_values = torch.tensor([
    [-2.5, 1.3, -0.8, 3.2, -1.1],
    [0.5, -2.0, 4.1, -3.5, 0.0]
])

print("输入张量（包含正数和负数）:")
print(test_values)
print()

# 应用 ReLU
relu = nn.ReLU()
output = relu(test_values)

print("ReLU 后的输出（负数变为0，正数保持）:")
print(output)
print()

# 展示对比
print("=== 对比演示 ===")
for i in range(test_values.shape[0]):
    for j in range(test_values.shape[1]):
        val_in = test_values[i, j].item()
        val_out = output[i, j].item()
        print(f"输入: {val_in:6.1f} → ReLU → 输出: {val_out:6.1f}")
    print()
````

# ReLU 的核心作用 / Key Benefits of ReLU

## 1️⃣ 引入非线性 / Introduces Non-linearity

```
✗ 没有激活函数：
  线性层 → 线性层 → 线性层 = 一个大的线性变换
  只能学习线性关系 (如：y = mx + b)

✓ 有 ReLU 激活：
  线性层 → [ReLU] → 线性层 → [ReLU] → 线性层
  能学习复杂的非线性关系
```

## 2️⃣ 促进稀疏性 / Promotes Sparsity

- 负值变为0（"关闭"这些神经元）
- 只让有用的正值通过
- 帮助模型**聚焦于重要特征**

## 3️⃣ 计算效率 / Computational Efficiency

- ReLU 计算非常简单：`if x > 0: return x else: return 0`
- 比其他激活函数（如 sigmoid、tanh）快得多
- 是现代深度学习的首选激活函数

## 在你的网络中的实际作用 / In Your Network

```python
hidden_output = hidden_layer(input)      # shape: (5, 15)
                                         # 可能有正值和负值

activated = activation(hidden_output)    # 应用 ReLU
                                         # 负值→0, 正值保留

output = output_layer(activated)         # 传给输出层
```

这样，**输出层接收的不是原始数据，而是经过ReLU筛选过的特征**，使模型更有表达力。

# 理解 torch.size() / Understanding torch.size()

## 什么是 `.size()`？

`.size()` 是一个方法，用来获取**张量的形状（维度信息）**。它返回一个 `torch.Size` 对象，包含张量的所有维度信息。

## 返回值示例

```python
tensor = torch.randn(5, 10)    # 创建形状为 (5, 10) 的张量
tensor.size()                   # 返回：torch.Size([5, 10])
```

## 理解返回的值

对于形状 `torch.Size([5, 10])`：

| 维度位置 | 值 | 含义 |
|---------|---|----|
| 第0维 | 5 | 有**5个样本** |
| 第1维 | 10 | 每个样本有**10个特征** |

## 在你的网络中的应用

```python
example_tensor = torch.randn(5, 10)
example_tensor.size()           # torch.Size([5, 10])

input_layer = nn.Linear(10, 20)
result = input_layer(example_tensor)
result.size()                   # torch.Size([5, 20])  ← 维度变了！
```

**数据流动过程：**
- 输入：5个样本，10个特征 → `[5, 10]`
- 通过 Linear(10→20)：保持样本数，改变特征维度 → `[5, 20]`
- 通过 Linear(20→15)：保持样本数，改变特征维度 → `[5, 15]`
- 通过 Linear(15→1)：保持样本数，改变特征维度 → `[5, 1]`

# 理解 SGD 优化器 / Understanding SGD Optimizer

## SGD 是什么？

**SGD (Stochastic Gradient Descent)** = **随机梯度下降**

它是一个**优化器**，用来在训练过程中自动**更新网络权重**，使损失函数逐步减小。

## 核心概念

### 1️⃣ 梯度下降的直观理解

想象你在一个山谷里，想要找到最低点：

```
        ↑ 损失
        |     ●  ← 当前位置（高损失）
        |    ╱╲
        |   ╱  ╲
        |  ╱    ╲  ← 下坡方向（负梯度方向）
        | ╱      ●
        |╱        ╲  ← 目标位置（最低损失）
        └──────────→ 权重
```

**梯度下降**：
- 计算当前位置的"下坡方向"（梯度）
- 沿着下坡方向走一小步
- 重复上述过程，最终到达山谷底部

### 2️⃣ SGD 中的关键参数

```python
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
                                          ↑              ↑
                                    学习率          动量参数
```

| 参数 | 含义 | 作用 |
|------|------|------|
| `model.parameters()` | 需要优化的权重 | 告诉优化器哪些参数需要更新 |
| `lr=0.001` | **学习率** | 每一步走多大的距离（0.001 = 很小的步长） |
| `momentum=0.9` | **动量** | 保持之前运动的方向，加速收敛 |

## 为什么叫"随机"梯度下降？

- **随机** = 每次使用一小批数据（batch）来计算梯度，而不是全部数据
- 这样更快、更节省内存、有时效果反而更好

## 在训练中的作用

```python
# 训练循环的关键步骤：
loss.backward()        # 计算梯度
optimizer.step()       # ← SGD 在这里工作！
                       # 更新权重：new_weight = old_weight - lr * gradient
```

SGD 根据计算出的梯度，自动朝着**降低损失**的方向调整每个权重。


````python
import torch
import torch.nn as nn
import torch.optim as optim

print("=== SGD 优化器的工作原理演示 ===\n")

# 创建一个简单的权重（为了演示）
simple_weight = torch.tensor(5.0, requires_grad=True)
print(f"初始权重值：{simple_weight.item()}")

# 创建 SGD 优化器
simple_optimizer = optim.SGD([simple_weight], lr=0.1)

print(f"\n优化器配置：")
print(f"  学习率 lr = 0.1")
print(f"  这意味着：每次更新时，权重改变 = -0.1 × 梯度\n")

# 模拟几步训练过程
print("模拟 5 步优化过程：\n")

for step in range(5):
    # 为了演示，构造一个简单的损失函数：loss = weight²
    loss = simple_weight ** 2
    
    print(f"步骤 {step + 1}:")
    print(f"  前向传播 → 损失 = {loss.item():.4f}")
    
    # 清零梯度
    simple_optimizer.zero_grad()
    
    # 反向传播计算梯度
    loss.backward()
    print(f"  反向传播 → 梯度 = {simple_weight.grad.item():.4f}")
    
    # 优化器更新权重 ← SGD 在这里工作！
    print(f"  SGD更新 → new_weight = {simple_weight.item():.4f} - 0.1 × {simple_weight.grad.item():.4f}")
    simple_optimizer.step()
    
    print(f"  更新后权重 = {simple_weight.item():.4f}\n")

print(f"最终权重值：{simple_weight.item():.6f}")
print(f"\n观察：权重逐步接近 0（最小值）！这就是梯度下降的效果。")
````

# SGD 在实际网络训练中的角色 / SGD's Role in Network Training

## 完整的训练流程

```python
# 0️⃣ 准备阶段
model = SimpleNeuralNetwork()
criterion = nn.CrossEntropyLoss()        # 损失函数
optimizer = optim.SGD(model.parameters(), lr=0.001)  # ← SGD在这里创建

# 1️⃣ 训练循环
for epoch in range(num_epochs):
    for inputs, labels in train_loader:
        
        # 前向传播
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        
        # 反向传播
        loss.backward()  # 计算所有权重的梯度
        
        # ✓ 优化 ← SGD在这一步更新所有权重！
        optimizer.step()
        
        # 清零梯度以准备下一次迭代
        optimizer.zero_grad()
```

## SGD 的三个关键动作

| 动作 | 代码 | 作用 |
|------|------|------|
| **计算梯度** | `loss.backward()` | 告诉优化器"权重应该朝哪个方向改变" |
| **更新权重** | `optimizer.step()` | SGD 实际执行：`weight = weight - lr × gradient` |
| **清零梯度** | `optimizer.zero_grad()` | 清除旧梯度，准备下一次计算 |

## 为什么需要 SGD？

✗ **不用优化器**：权重永远不会改变，模型无法学习
✓ **用 SGD**：权重不断调整，模型逐步改进，损失降低

## 学习率的影响

```
lr 太大        ↗↗↗↗  震荡，可能发散
              ╱
lr 合适        ╱  平稳收敛 ✓
             ╱
lr 太小        ╱ 收敛太慢
```

在你的代码中：`lr=0.001` 是一个比较保守的值，适合大多数情况。


````python
import torch
import torch.optim as optim

print("=== 不同学习率的 SGD 对比 ===\n")

# 测试函数：loss = (weight - 3)²，最优权重是 3
def create_sgd_test(initial_weight, lr, name):
    weight = torch.tensor(initial_weight, requires_grad=True)
    optimizer = optim.SGD([weight], lr=lr)
    
    print(f"{name} (学习率={lr}):")
    print(f"  初始权重: {weight.item():.4f}")
    
    # 进行 10 步优化
    for _ in range(10):
        loss = (weight - 3) ** 2
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    print(f"  10步后权重: {weight.item():.4f}")
    print(f"  目标是 3.0，距离: {abs(weight.item() - 3):.4f}\n")

# 对比不同学习率
create_sgd_test(0.0, lr=0.01, name="学习率小")
create_sgd_test(0.0, lr=0.1,  name="学习率中")
create_sgd_test(0.0, lr=0.5,  name="学习率大")
create_sgd_test(0.0, lr=1.0,  name="学习率太大（易发散）")

print("=" * 50)
print("总结：")
print("  ✓ 学习率太小：收敛慢")
print("  ✓ 学习率合适：快速接近目标")
print("  ✗ 学习率太大：可能无法收敛或发散")
````

# SGD 和其他优化器的对比 / SGD vs Other Optimizers

## 常见优化器

| 优化器 | 特点 | 适用场景 |
|--------|------|---------|
| **SGD** | 简单、稳定 | 大多数场景，特别是CNN |
| **Adam** | 自适应学习率，收敛快 | 现代深度学习（推荐） |
| **RMSprop** | 自适应学习率 | RNN等 |
| **Momentum** | SGD + 动量加速 | 当 SGD 卡住时 |

## 你在代码中的 SGD 配置

```python
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

- `lr=0.001`：每步移动距离很小，稳定但可能慢
- `momentum=0.9`：强动量，像一个有"惯性"的球滚下坡

## 关键总结 🎯

1. **SGD 的作用**：在训练过程中自动调整网络权重
2. **如何工作**：梯度指向 → SGD 朝着反方向走 → 损失下降
3. **学习率很重要**：太小会慢，太大会发散
4. **每个 epoch 运行多次**：对每个批次的数据运行一次 `optimizer.step()`

## 记住这个流程

```
训练数据 → 模型 → 损失 → 反向传播(计算梯度)
                         ↓
                    SGD更新权重 ← 这就是学习！
                         ↓
                    重复，损失逐步下降
```

**这就是深度学习的核心：通过不断地计算梯度和更新权重，让网络自动学习如何做出更好的预测！**

# Create a NN for an Image Classifier / 为图像分类器创建神经网络
Here we create a neural network to train an image classification model. / 这里我们将创建一个神经网络来训练图像分类模型。



````python
# Create the class / 创建网络类
import torch.nn as nn
# This module simplifies importing operations (activation functions) / 该模块便于导入操作（激活函数）
import torch.nn.functional as F 

class ImageClassificationNet(nn.Module):
    def __init__(self):
        super(ImageClassificationNet, self).__init__()
        # Takes 1 input channel, outputs 6 feature maps, uses a 5x5 kernel
# 输入1通道，输出6个特征图，卷积核为5x5
        self.conv1 = nn.Conv2d(1, 6, 5)
        # Takes 6 input feature maps, outputs 16 feature maps, uses a 5x5 kernel
# 输入6个特征图，输出16个特征图，卷积核为5x5
        self.conv2 = nn.Conv2d(6, 16, 5)
        # Define a max pooling layer to downsample feature maps by a factor of 2
# 定义最大池化层，将特征图下采样2倍
        self.pool = nn.MaxPool2d(2, 2)
        # Flattened conv output -> 120 units / 卷积展平输出 -> 120单元
        self.fc1 = nn.Linear(16 * 4 * 4, 120)
        # Second fully connected layer: 120 -> 84 / 第二全连接层：120 -> 84
        self.fc2 = nn.Linear(120, 84)
        # 10 classes for classification / 分类任务共10类
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        # conv1 + ReLU + maxpool / 第1层卷积 + ReLU + 最大池化
        x = self.pool(F.relu(self.conv1(x)))
        # conv2 + ReLU + maxpool / 第2层卷积 + ReLU + 最大池化
        x = self.pool(F.relu(self.conv2(x)))
        # Flatten while keeping batch dimension / 展平并保留批次维度
        x = torch.flatten(x, 1)
        # Fully connected + ReLU / 全连接 + ReLU
        x = F.relu(self.fc1(x))
        # Fully connected + ReLU / 全连接 + ReLU
        x = F.relu(self.fc2(x))
        # Output logits / 输出logits
        x = self.fc3(x)
        # Return class scores / 返回类别分数
        return x
````

# ImageClassificationNet 模型结构详解

## 模型的全貌

这个模型称为 **ImageClassificationNet**，是一个**卷积神经网络（CNN）**，用于将**图像分类为10个类别**。

## 模型的两个部分

### 第一部分：特征提取层（卷积层 + 池化层）

```
输入图像 (1×28×28)  [灰度图，28×28像素]
    ↓
[Conv2d] 1→6 通道，5×5卷积核
    ↓
[ReLU] 激活函数
    ↓
[MaxPool] 2×2 最大池化，图像缩小2倍
    ↓
[Conv2d] 6→16 通道，5×5卷积核
    ↓
[ReLU] 激活函数
    ↓
[MaxPool] 2×2 最大池化，图像再缩小2倍
    ↓
特征向量 (16×4×4=256个特征)
```

### 第二部分：分类层（全连接层）

```
特征向量 (256)
    ↓
[Linear] 256→120 全连接层
    ↓
[ReLU] 激活函数
    ↓
[Linear] 120→84 全连接层
    ↓
[ReLU] 激活函数
    ↓
[Linear] 84→10 全连接层（输出10个类别分数）
    ↓
最终预测 (10个类别)
```

## 各层的具体作用

| 层名 | 类型 | 作用 | 输入形状 | 输出形状 |
|------|------|------|---------|---------|
| conv1 | Conv2d(1,6,5) | 提取基础特征（边缘、纹理）| (batch, 1, 28, 28) | (batch, 6, 24, 24) |
| pool | MaxPool2d(2,2) | 减少数据量，保留重要信息 | (batch, 6, 24, 24) | (batch, 6, 12, 12) |
| conv2 | Conv2d(6,16,5) | 提取高层特征（形状、组成） | (batch, 6, 12, 12) | (batch, 16, 8, 8) |
| pool | MaxPool2d(2,2) | 再次降采样 | (batch, 16, 8, 8) | (batch, 16, 4, 4) |
| flatten | - | 展平为1维向量 | (batch, 16, 4, 4) | (batch, 256) |
| fc1 | Linear(256, 120) | 合并特征，学习特征组合 | (batch, 256) | (batch, 120) |
| fc2 | Linear(120, 84) | 进一步提炼特征 | (batch, 120) | (batch, 84) |
| fc3 | Linear(84, 10) | 输出10个类别的分数 | (batch, 84) | (batch, 10) |

## 卷积层和池化层是什么？

### Conv2d（卷积层）

```
卷积就像用一个小窗口（卷积核）在图像上滑动，提取局部特征：

原始图像：          卷积核（5×5）：
█ █ █ █ █          ■ ■ ■ ■ ■
█ █ █ █ █          ■ ■ ■ ■ ■
█ █ █ █ █           ■ ■ ■ ■ ■
█ █ █ █ █          ■ ■ ■ ■ ■
█ █ █ █ █          ■ ■ ■ ■ ■
                      ↓
                   检测边缘、纹理等特征
```

### MaxPool2d（最大池化）

```
原始特征图：        最大池化2×2：
1 2 3 4            |最大值：2|最大值：4|
5 6 7 8      →     |---------|---------|
9 10 11 12         |最大值：10|最大值：12|
13 14 15 16        |---------|---------|

作用：
✓ 减少数据量（从16变成4个值）
✓ 保留最重要的特征
✓ 加快计算速度
```

## 为什么是这个顺序？

```
卷积 + ReLU + 池化    → 逐层提取和压缩图像信息
       ↓
展平（Flatten）      → 将2D特征转换为1D向量
       ↓
全连接层 + ReLU      → 学习特征间的关系
       ↓
输出层（10个类）     → 给出分类结果
```

## 这个特定配置为什么这样设计？

- **1个输入通道**：灰度图像（黑白）
- **输出10个类别**：FashionMNIST 有10种衣服类型（T恤、裤子、靴子等）
- **5×5卷积核**：足够大来捕捉小的特征
- **两层卷积**：第一层学习简单特征，第二层学习复杂特征
- **池化**：减少过拟合，加快训练
- **三层全连接**：逐步将特征转换为分类决策


````python
import torch
import torch.nn as nn
import torch.nn.functional as F

print("=== 数据在 ImageClassificationNet 中的流动 ===\n")

# 创建一个示例批次（4张图像，每张28×28的灰度图）
batch_size = 4
sample_input = torch.randn(batch_size, 1, 28, 28)
print(f"输入图像批次 (4张图像)：{sample_input.shape}")
print(f"  → {batch_size} 张图像")
print(f"  → 每张 1 个通道（灰度）")
print(f"  → 每张 28×28 像素\n")

# 手动演示网络的每一层
print("=" * 60)
print("网络各层的形状变化：")
print("=" * 60 + "\n")

# Conv1
conv1 = nn.Conv2d(1, 6, 5)
x = conv1(sample_input)
print(f"1️⃣  [Conv2d(1→6, kernel=5)]")
print(f"    输入:  {sample_input.shape}")
print(f"    输出:  {x.shape}")
print(f"    说明:  1 个通道变 6 个通道，28×28 变 24×24\n")

# ReLU + Pool
x = F.relu(x)
pool = nn.MaxPool2d(2, 2)
x = pool(x)
print(f"2️⃣  [ReLU + MaxPool(2,2)]")
print(f"    输入:  torch.Size([{batch_size}, 6, 24, 24])")
print(f"    输出:  {x.shape}")
print(f"    说明:  ReLU激活，池化将 24×24 变 12×12\n")

# Conv2
conv2 = nn.Conv2d(6, 16, 5)
x = conv2(x)
print(f"3️⃣  [Conv2d(6→16, kernel=5)]")
print(f"    输入:  torch.Size([{batch_size}, 6, 12, 12])")
print(f"    输出:  {x.shape}")
print(f"    说明:  6 个通道变 16 个通道，12×12 变 8×8\n")

# ReLU + Pool
x = F.relu(x)
x = pool(x)
print(f"4️⃣  [ReLU + MaxPool(2,2)]")
print(f"    输入:  torch.Size([{batch_size}, 16, 8, 8])")
print(f"    输出:  {x.shape}")
print(f"    说明:  激活，池化将 8×8 变 4×4\n")

# 展平
x_flat = torch.flatten(x, 1)
print(f"5️⃣  [Flatten]")
print(f"    输入:  {x.shape}")
print(f"    输出:  {x_flat.shape}")
print(f"    说明:  将 (4, 16, 4, 4) 展平为 (4, 256)")
print(f"           其中 256 = 16 × 4 × 4\n")

# Fully Connected 1
fc1 = nn.Linear(16 * 4 * 4, 120)
x = fc1(x_flat)
print(f"6️⃣  [Linear(256→120)]")
print(f"    输入:  {x_flat.shape}")
print(f"    输出:  {x.shape}")
print(f"    说明:  全连接层，256 个特征→120 个特征\n")

# Fully Connected 2
x = F.relu(x)
fc2 = nn.Linear(120, 84)
x = fc2(x)
print(f"7️⃣  [ReLU + Linear(120→84)]")
print(f"    输入:  torch.Size([{batch_size}, 120])")
print(f"    输出:  {x.shape}")
print(f"    说明:  激活，全连接层，120→84\n")

# Output
x = F.relu(x)
fc3 = nn.Linear(84, 10)
x = fc3(x)
print(f"8️⃣  [ReLU + Linear(84→10)]")
print(f"    输入:  torch.Size([{batch_size}, 84])")
print(f"    输出:  {x.shape}")
print(f"    说明:  最终输出，10 个类别分数\n")

print("=" * 60)
print("总结：")
print(f"  输入：  (4, 1, 28, 28)   → 4张灰度图像")
print(f"  输出：  (4, 10)           → 4张图像的分类结果")
print(f"         每个图像有10个类别分数，代表属于各类的概率")
print("=" * 60)
````

# 模型结构为什么这样设计？

## 🧠 任务：图像分类

**输入**：一张衣服的图像（28×28像素，灰度）
**输出**：预测这件衣服是什么类型（10种类型：T恤、裤子、靴子等）

## 📊 网络的两个主要阶段

### 阶段1：特征提取（卷积 + 池化）

层数少，但很深层（28×28→24→12→8→4）：

```
Conv1 (1→6)     学习最基本的特征
    ↓
ReLU + Pool     保留强信号，丢弃弱信号
    ↓
Conv2 (6→16)    学习更复杂的特征组合
    ↓
ReLU + Pool     再次过滤和压缩
```

**为什么这样？**
- **卷积**：保留像素的**空间关系**（边相邻）
- **池化**：减少信息量，防止过拟合
- **通道增加**（1→6→16）：学习更丰富的特征表示

### 阶段2：分类决策（全连接层）

```
展平向量 (256个数字)
    ↓
FC1 (256→120)   学习特征的抽象组合
    ↓
FC2 (120→84)    进一步提炼
    ↓
FC3 (84→10)     输出10个类别分数
```

**为什么这样？**
- **全连接层**：学习不同特征之间的**关系**
- **逐层缩小**（256→120→84→10）：从粗到细，逐步做出决策

## 对比：为什么不能是其他设计？

```
如果只用全连接层：               如果太多卷积层：
输入 (1×28×28)                  容易过拟合
  ↓                              计算量太大
fc1 (784→120)  ✗ 丢失空间信息      训练太慢
fc2 (120→10)                     ✗ 不必要

如果太浅：
Conv1 - fc1 - fc3  ✗ 信息丢失太多
无法学到复杂特征
```

## 🎯 核心思想

```
原始图像信息 (高维)
    ↓↓↓
逐步提取重要特征 (中间维度)
    ↓↓↓
最终做出分类决策 (低维：10个类别)
```

这个设计遵循了**CNN的黄金法则**：
- **前面**：用卷积提取**局部特征**（快速，有效）
- **后面**：用全连接做**全局决策**（灵活，精准）


````python
import torch
from torch import nn

print("=== ImageClassificationNet 快速参考 ===\n")

# 创建模型来展示
class ImageClassificationNet(nn.Module):
    def __init__(self):
        super(ImageClassificationNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, 5)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(16 * 4 * 4, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = torch.flatten(x, 1)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

model = ImageClassificationNet()

print("一句话总结：")
print("  这是一个 CNN 分类器，用6层从图像中提取特征，")
print("  最后输出10个类别中每个的得分。\n")

print("模型参数统计：")
total_params = sum(p.numel() for p in model.parameters())
print(f"  总参数数：{total_params:,} 个\n")

print("每层参数数：")
for name, param in model.named_parameters():
    num_params = param.numel()
    print(f"  {name:15} : {num_params:8,} 个参数  形状: {tuple(param.shape)}")

print("\n数据流向图：")
print("""
  ┌─────────────────────────────────────────────────────┐
  │ 输入图像 (1×28×28)                                   │
  └──────────────────┬──────────────────────────────────┘
                     ↓
  ┌─────────────────────────────────────────────────────┐
  │ 【特征提取阶段】                                      │
  │ ├─ Conv2d(1→6, k=5) : 1×28×28 → 6×24×24              │
  │ ├─ MaxPool(2×2)     : 6×24×24 → 6×12×12              │
  │ ├─ Conv2d(6→16,k=5) : 6×12×12 → 16×8×8               │
  │ └─ MaxPool(2×2)     : 16×8×8 → 16×4×4                │
  │ 输出：16×4×4 = 256 个特征                             │
  └──────────────────┬──────────────────────────────────┘
                     ↓
  ┌─────────────────────────────────────────────────────┐
  │ 【分类决策阶段】                                      │
  │ ├─ Flatten        : 256 个数字                       │
  │ ├─ Linear(256→120): 120 个神经元                     │
  │ ├─ Linear(120→84) : 84 个神经元                      │
  │ └─ Linear(84→10)  : 10 个类别                        │
  │ 输出：10个分数 [0.1, 0.8, 0.05, ...]                 │
  └──────────────────┬──────────────────────────────────┘
                     ↓
  ┌─────────────────────────────────────────────────────┐
  │ 预测结果：最高分的类别就是预测                        │
  │ argmax([0.1, 0.8, 0.05, ...]) = 索引 1               │
  │ → 预测是"裤子" (FashionMNIST的第二类)                │
  └─────────────────────────────────────────────────────┘
""")
````

# Create a Training Loop / 创建训练循环
A training loop in PyTorch repeatedly feeds data through the model, computes loss, and updates parameters to minimize loss. / 在 PyTorch 中，训练循环会反复将数据送入模型、计算损失并更新参数，以最小化损失。

This continues for a fixed number of epochs or until performance is satisfactory. / 这个过程会持续固定轮数（epochs），或直到模型性能达到预期。

````python
# Define number of epochs / 定义训练轮数
N_EPOCHS = 3
````

````python
for epoch in range(N_EPOCHS):  # Loop over the dataset N_EPOCH times / 按轮次遍历数据集
    
    running_loss = 0.0  # Initialize running loss for current epoch / 初始化当前轮累计损失
    
    # Loop over training data in batches / 按批次遍历训练数据
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data  # Unpack data: inputs are images, labels are classes / 解包数据：inputs为图像，labels为类别
        
        optimizer.zero_grad()  # Clear gradients to avoid accumulation / 清空梯度，避免累加

        outputs = model(inputs)  # Forward pass: compute predictions / 前向传播：计算预测
        loss = criterion(outputs, labels)  # Compute loss / 计算损失
        loss.backward()  # Backward pass: compute gradients / 反向传播：计算梯度
        optimizer.step()  # Update model parameters / 更新模型参数
        
        running_loss += loss.item()  # Accumulate loss / 累加损失

    # Print average loss for this epoch / 打印当前轮平均损失
    print(f"Epoch / 轮次: {epoch} Loss / 损失: {running_loss/len(train_loader)}")
````

# Create a Training Loop with Validation / 创建带验证的训练循环

````python
for epoch in range(N_EPOCHS):  # Loop over the dataset N_EPOCH times / 按轮次遍历数据集
    
    ####### TRAINING / 训练阶段
    training_loss = 0.0  # Initialize training loss / 初始化训练损失
    # Set model to training mode / 将模型设置为训练模式
    model.train()
    # Loop over training data in batches / 按批次遍历训练数据
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data  # Unpack data / 解包数据
        
        optimizer.zero_grad()  # Clear gradients / 清空梯度

        outputs = model(inputs)  # Compute predictions / 计算预测输出
        loss = criterion(outputs, labels)  # Compute loss / 计算损失
        loss.backward()  # Compute gradients / 计算梯度
        optimizer.step()  # Update parameters / 更新参数
        
        training_loss += loss.item()  # Accumulate training loss / 累加训练损失

    ######## VALIDATION / 验证阶段
    val_loss = 0.0 # Initialize validation loss / 初始化验证损失
    # Set model to evaluation mode / 将模型设置为评估模式
    model.eval()

    # Loop over validation data in batches / 按批次遍历验证数据
    for i, data in enumerate(val_loader, 0):
        inputs, labels = data  # Unpack validation data / 解包验证数据
        
        outputs = model(inputs)  # Compute predictions / 计算预测
        loss = criterion(outputs, labels)  # Compute loss / 计算损失
        
        #### NOTICE: no gradient computation or weight updates #### 
        #### 注意：此处不计算梯度，也不更新权重 ####
        val_loss += loss.item()  # Accumulate validation loss / 累加验证损失

    # Print training and validation loss / 打印训练与验证损失
    print(f"Epoch / 轮次: {epoch} Train Loss / 训练损失: {training_loss/len(train_loader)} Val Loss / 验证损失: {val_loss/len(val_loader)}")
````

# About Loss / 关于损失
If validation loss keeps decreasing, the model is generally learning well. / 如果验证损失持续下降，通常说明模型学习良好。

If training loss decreases but validation loss does not, the model is likely overfitting. / 如果训练损失下降而验证损失不降，通常说明模型可能过拟合。

In practice, we may need more epochs and further tuning for better accuracy. / 在实践中，通常还需要更多训练轮次和进一步调参来提升准确率。
