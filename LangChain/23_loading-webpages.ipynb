```python
# ============================================================
# Block 1: Import Dependencies and Load Environment
# 模块1：导入依赖库并加载环境变量
# ============================================================
# Import environment management library
# Best practice: Always maintain environment configuration injection
# 导入环境管理库
# 最佳实践：始终保持环境配置注入
import os
from dotenv import load_dotenv

# Load environment variables
# 加载环境变量
load_dotenv(override=True)


# ============================================================
# Block 2: Import LangChain Web Loader
# 模块2：导入LangChain网页加载器
# ============================================================
# Import web-based document loader for scraping web content
# 导入基于Web的文档加载器，用于抓取网页内容
from langchain_community.document_loaders import WebBaseLoader


# ============================================================
# Block 3: Define Target Data Source
# 模块3：定义目标数据源
# ============================================================
# Target URL: The Verge article about Meta AI Assistant
# This is a real-world example demonstrating web scraping capabilities
# 目标URL：The Verge关于Meta AI助手的文章
# 这是一个真实案例，演示网页抓取能力
URL = "https://www.theverge.com/2024/4/18/24133808/meta-ai-assistant-llama-3-chatgpt-openai-rival"


# ============================================================
# Block 4: Load and Parse Webpage (Web Scraping & Parsing)
# 模块4：加载并解析网页（网页抓取与解析）
# ============================================================
print("=" * 70)
print("Web Scraping in Progress")
print("=" * 70)
print(f"Target URL: {URL}")
print()

# Instantiate the web loader with target URL
# 使用目标URL实例化网页加载器
loader = WebBaseLoader(URL)

# Execute loading - this performs HTTP request, HTML parsing, and text extraction
# Returns a list of Document objects (typically length 1 for single webpage)
# 执行加载 - 这会执行HTTP请求、HTML解析和文本提取
# 返回Document对象列表（单个网页通常长度为1）
data = loader.load()

print("✅ Webpage Loading Completed Successfully!")
print()


# ============================================================
# Block 5: Validate Single-Element Property and Metadata
# 模块5：验证单元素特性和元数据
# ============================================================
# KEY OBSERVATION: Web pages are loaded as a single Document object
# Unlike PDFs which split by page, web content is typically returned as one chunk
# 关键观察：网页被加载为单个Document对象
# 与按页分割的PDF不同，网页内容通常作为一个块返回

print("=" * 70)
print("Document Analysis / 文档分析")
print("=" * 70)
print(f"Number of Document Objects Extracted: {len(data)}")
print(f"提取的Document对象数量: {len(data)}")
print()

# Extract the single document object
# 提取唯一的文档对象
web_doc = data[0]


# --------------------------------------------------------
# Sub-block 5.1: Inspect Automatically Extracted Metadata
# 子模块5.1：检查自动提取的元数据
# --------------------------------------------------------
print("-" * 70)
print("Automatically Extracted Metadata / 自动提取的元数据")
print("-" * 70)

# WebBaseLoader intelligently extracts metadata like title, description, source
# This is crucial for citation tracking and source verification
# WebBaseLoader智能提取元数据，如标题、描述、来源
# 这对引用跟踪和来源验证至关重要
for key, value in web_doc.metadata.items():
    print(f"- {key.capitalize()}: {value}")
print()


# --------------------------------------------------------
# Sub-block 5.2: Preview Cleaned Plain Text Content
# 子模块5.2：预览清理后的纯文本内容
# --------------------------------------------------------
print("-" * 70)
print("Content Preview (First 500 Characters)")
print("-" * 70)

# .strip() removes leading/trailing whitespace and newlines
# This is essential for cleaning real-world messy HTML data
# .strip() 去除首尾空白符和换行符
# 这对清理真实世界中杂乱的HTML数据至关重要
preview_text = web_doc.page_content.strip()[:500]
print(preview_text)

total_length = len(web_doc.page_content.strip())
if total_length > 500:
    print("\n... (truncated, showing first 500 chars)")
print()


# --------------------------------------------------------
# Sub-block 5.3: Display Content Statistics
# 子模块5.3：显示内容统计
# --------------------------------------------------------
print("-" * 70)
print("Content Statistics / 内容统计")
print("-" * 70)
print(f"Total Characters: {len(web_doc.page_content)}")
print(f"Total Words (approx): {len(web_doc.page_content.split())}")
print()

print("=" * 70)
print("Web Scraping Completed Successfully")
print("=" * 70)
```
