```python

# ============================================================================
# [部分 1] 环境初始化与依赖导入 / Part 1: Environment Setup and Imports
# ============================================================================
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
# 直接导入"消息类"而不是 Prompt Template
# Import Message classes directly instead of using PromptTemplate
from langchain_core.messages import HumanMessage, SystemMessage

# 强制加载环境变量（包含 OpenAI API Key）
# Force load environment variables (containing OpenAI API Key)
load_dotenv(override=True)


# ============================================================================
# [部分 2] 实例化语言模型 / Part 2: Initialize Language Model
# ============================================================================

# 创建 ChatOpenAI 实例，temperature=0 确保严谨的输出
# Create ChatOpenAI instance, temperature=0 ensures deterministic output
# 这里的 model 对象是一个"通用的聊天接口"，可以接收各种格式的输入
# The model object is a "universal chat interface" that accepts various input formats
model = ChatOpenAI(model="gpt-4o-mini", temperature=0)


# ============================================================================
# [部分 3] 定义系统消息（AI 的角色和行为准则）/ Part 3: Define System Message
# ============================================================================

# SystemMessage：告诉 AI 它应该是什么角色、如何表现
# SystemMessage: tells AI what role it should play and how to behave
# 这是整个"对话上下文"的基础。所有后续的回复都会遵循这个角色定义
# This is the foundation of the entire "conversation context". 
# All subsequent responses will follow this role definition
sysmsg = "You are a Physics teacher"


# ============================================================================
# [部分 4] 定义人类消息（用户的真实问题）/ Part 4: Define Human Message
# ============================================================================

# HumanMessage：用户真正要问的问题、要求执行的任务
# HumanMessage: the actual question the user is asking or task to perform
# 对比 ChatPromptTemplate：这里没有占位符，直接就是完整的问题
# Unlike ChatPromptTemplate: no placeholders here, just the complete question
humanmsg = "Explain the concept of Galaxy in simple terms."


# ============================================================================
# [部分 5] 将消息打包成列表 / Part 5: Package Messages into List
# ============================================================================

# LangChain 的消息系统要求把所有消息放在一个列表中
# LangChain's message system requires all messages to be in a single list
# 消息顺序很重要：通常 SystemMessage 在前，HumanMessage 在后
# Message order matters: typically SystemMessage first, then HumanMessage
# 这个列表就是要发送给 LLM API 的完整"对话历史"
# This list is the complete "conversation history" to send to the LLM API
messages = [SystemMessage(content=sysmsg), HumanMessage(content=humanmsg)]


# ============================================================================
# [部分 6] 调用模型并获取响应 / Part 6: Invoke Model and Get Response
# ============================================================================

print("Waiting for the physics teacher's answer...\n")

# 直接调用 model.invoke()，传入消息列表（而不是 Prompt 对象）
# Directly call model.invoke() with message list (not a Prompt object)
# LLM 会：
#   1. 看到 SystemMessage，理解自己是"物理老师"
#   2. 看到 HumanMessage，获得用户的问题
#   3. 根据这两个消息生成对应的回复
#
# The LLM will:
#   1. See SystemMessage and understand it's a "Physics teacher"
#   2. See HumanMessage and get the user's question
#   3. Generate appropriate response based on both messages
response = model.invoke(messages)


# ============================================================================
# [部分 7] 打印和处理响应 / Part 7: Display Response
# ============================================================================

# response 是一个 AIMessage 对象（由 LLM 生成）
# response is an AIMessage object (generated by the LLM)
# AIMessage 的 .content 属性包含了实际的文本内容
# The .content attribute of AIMessage contains the actual text
print(response.content)
```
