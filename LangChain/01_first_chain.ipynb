```python

# ============================================================================
# [部分 1] 环境初始化和依赖导入 / Part 1: Environment Setup and Dependencies
# ============================================================================
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

# 加载环境变量中的 API 秘钥 / Load API key from environment variables
load_dotenv(override=True)


# ============================================================================
# [部分 2] 实例化大语言模型 / Part 2: Initialize Language Model (LLM)
# ============================================================================

# 创建 ChatOpenAI 实例，指定使用 gpt-4o-mini 模型
# Create a ChatOpenAI instance with gpt-4o-mini model
# 这个对象是 LangChain 中的 Language Model，可以直接与 OpenAI API 通讯
# This is a Language Model component that communicates with OpenAI API
llm = ChatOpenAI(model="gpt-4o-mini")


# ============================================================================
# [部分 3] 构建提示词模板 / Part 3: Create Prompt Template
# ============================================================================

# 使用 ChatPromptTemplate 构建一个包含 system 和 user 消息的提示词模板
# Create a ChatPromptTemplate with system and user message roles
# - system: 定义 AI 的角色和行为准则 / Define AI's role and behavior guidelines
# - user: 动态的用户输入，其中 {paper_summary} 是一个占位符 / Dynamic user input with {paper_summary} placeholder
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a top-tier Natural Language Processing (NLP) research assistant. Please extract information strictly in accordance with the requirements."),
    ("user", "Please summarize the core contribution of this paper in one sentence: {paper_summary}")
])


# ============================================================================
# [部分 4] 构建处理链 / Part 4: Create the Processing Chain (Core of LangChain)
# ============================================================================

# LangChain 的核心魔法：使用 | (pipe) 操作符将不同组件链接起来
# The magic of LangChain: pipe (|) operator chains components together
# 流程：用户输入 → 提示词模板格式化 → 发送给 LLM → 得到 LLM 的响应
# Flow: User input → Prompt template formats → Send to LLM → Get LLM response
chain = prompt | llm


# ============================================================================
# [部分 5] 执行链并处理结果 / Part 5: Execute Chain with Dynamic Variables
# ============================================================================

# 使用 invoke() 方法运行链，传入动态变量替换 {paper_summary} 占位符
# Use invoke() to run the chain, substituting the {paper_summary} placeholder
# 字典中的 "paper_summary" 值会被替换到模板中的 {paper_summary} 位置
# The "paper_summary" value replaces the {paper_summary} placeholder in the template
response = chain.invoke({
    "paper_summary": "This paper proposes a new architecture called RAG, which provides context for large language models by introducing an external vector database, thereby significantly reducing the hallucination rate of the models in professional domain question answering and improving the accuracy of information extraction."
})


# ============================================================================
# [部分 6] 输出结果 / Part 6: Display Results
# ============================================================================

# 响应对象的 .content 属性包含了 LLM 生成的文本内容
# The .content property of the response contains the text generated by LLM
print(response.content)
```
