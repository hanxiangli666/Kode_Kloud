```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

# 1. 加载你的 OpenAI 秘钥
load_dotenv(override=True)

# 2. 实例化语言模型 (这是你刚才学到的 Language Model 模块)
# 这里的代码完全解耦，以后你可以把 ChatOpenAI 直接换成 ChatOllama 来调用本地模型！
llm = ChatOpenAI(model="gpt-4o-mini")

# 3. 实例化提示词模板 (这是你刚才学到的 Prompt 模块)
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个顶级的自然语言处理(NLP)科研助手。请严格按照要求提取信息。"),
    ("user", "请用一句话概括这篇论文的核心贡献：{paper_summary}")
])

# 4. 见证 LangChain 的魔法：将 Prompt 和 LLM 链接(Chain)在一起
chain = prompt | llm

# 5. 运行链条并传入动态变量 (未来这里的文本可以是你用代码从 PDF 里读出来的)
response = chain.invoke({
    "paper_summary": "本文提出了一种名为 RAG 的新型架构，通过引入外部向量数据库来为大语言模型提供上下文，从而显著降低了模型在专业领域问答中的幻觉率，并提高了信息抽取的准确度。"
})

# 6. 打印结果
print(response.content)
```
