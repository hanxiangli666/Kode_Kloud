```python
import os
# 【硬性要求】加载环境变量 (科研级项目的标准起手式)
from dotenv import load_dotenv
load_dotenv(override=True)

# 导入垂直领域的雅虎财经新闻工具
from langchain_community.tools.yahoo_finance_news import YahooFinanceNewsTool

# ==========================================
# 1. 实例化工具 (Tool Initialization)
# ==========================================
# 这个工具不需要配置 API Key，它底层依赖于开源的 yfinance 爬虫库
finance_tool = YahooFinanceNewsTool()

print("=== 🛠️ Exploring Financial Instrument Metadata ===")
print(f"Name: {finance_tool.name}")
print(f"Description: {finance_tool.description}")

# ==========================================
# 2. 执行垂直定向查询 (Executing Domain-Specific Query)
# ==========================================
if __name__ == "__main__":
    # 设定目标资产的股票代码 (Ticker Symbol)，老师用的是英伟达 NVDA
    target_ticker = "GOOGL"
    print(f"\n=== 📈 Currently capturing the latest financial developments of {target_ticker} ===")
    
    # 【科研级保姆注释】
    # 旧版本写法：res = tool.run("NVDA")
    # 现代工程写法：必须使用 .invoke()。向工具传递一个字典，键通常根据工具的 args_schema 而定。
    # 对于大多数简单工具，传递一个包含单一字符串或明确参数字典的对象即可。
    try:
        # 使用现代 invoke 调用
        response = finance_tool.invoke(target_ticker)
        
        print("\n=== 📄 返回的新闻摘要 (Context ready for LLM) ===")
        # 输出预期：你将看到类似 "Nvidia (NVDA) closed at $XXX..." 的实时财经快讯
        print(response)
        
    except Exception as e:
        print(f"\n❌ 执行失败，请确保你已安装 yfinance (pip install yfinance): {e}")
```
