```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# 1. 激活并读取当前目录下的 .env 文件
load_dotenv(override=True)  # override=True 允许覆盖环境变量

# 2. 现代版语法：初始化客户端
client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

# 3. 发送时区测试请求
response = client.chat.completions.create(
    model="gpt-4o", 
    messages=[
        {"role": "system", "content": "You are a helpful AI assistant that answer questions related to timezones"},
        {"role": "user", "content": "If it is 9AM in New York, what time is it in China Shenzhen. Be concise."}
    ]
)

# 4. 精准打印 AI 回复的内容
print(response.choices[0].message.content)
```
