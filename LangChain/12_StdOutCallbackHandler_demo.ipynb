```python
import os
from dotenv import load_dotenv

# 1. ç°ä»£åŒ–çš„å¯¼åŒ…
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
# å¼•å…¥æ ¸å¿ƒçš„å›è°ƒå¤„ç†å™¨ï¼ˆç”¨æ¥å¹³æ›¿æ—§ç‰ˆçš„ verbose=Trueï¼‰
from langchain_core.callbacks import StdOutCallbackHandler 

# 2. å¿…å¤‡èµ·æ‰‹å¼ï¼šå¼ºåˆ¶åŠ è½½æœ¬åœ° API å¯†é’¥
load_dotenv(override=True)

# 3. å®šä¹‰æç¤ºè¯æ¨¡æ¿ (ChatPromptTemplate)
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a {subject} teacher"),
    ("human", "Tell me about {concept}")
])

# 4. åˆå§‹åŒ–å¤§è¯­è¨€æ¨¡å‹ (LLM)
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 5. åˆå§‹åŒ–çº¯æ–‡æœ¬è§£æå™¨ (å‰¥ç¦»å¤æ‚çš„å¯¹è±¡å¤–å£³)
parser = StrOutputParser()

# =================================================================
# ğŸš€ å½»åº•æ·˜æ±° LLMChainï¼Œç»„è£…æå…¶ä¼˜é›…çš„ç°ä»£ LCEL æµæ°´çº¿
# =================================================================
chain = prompt | llm | parser

print("========== å¼€å¯è¯¦ç»†æ—¥å¿—ç›‘æ§ (å¹³æ›¿ verbose=True) ==========\n")

# 6. å®ä¾‹åŒ–å›è°ƒå¤„ç†å™¨
handler = StdOutCallbackHandler()

# 7. ç°ä»£çš„è°ƒç”¨æ–¹å¼ï¼šé€šè¿‡ config æ³¨å…¥å›è°ƒç›‘æ§ï¼
# è¿™æ ·ä¸ä»…èƒ½çœ‹åˆ°æœ€ç»ˆç»“æœï¼Œè¿˜èƒ½åƒ Xå…‰æœºä¸€æ ·çœ‹åˆ°æ•°æ®æ˜¯æ€ä¹ˆæµè¿›æç¤ºè¯çš„
response = chain.invoke(
    {"subject": "physics", "concept": "galaxy"},
    config={'callbacks': [handler]}  # ğŸŒŸ æ ¸å¿ƒé­”æ³•åœ¨è¿™é‡Œï¼
)

print("\n========== æœ€ç»ˆå¹²å‡€çš„è¾“å‡ºæ–‡æœ¬ ==========\n")
print(response)
```
