```python
# ============================================================
# Block 1: Import Dependencies
# 模块1：导入依赖库
# ============================================================
# Import necessary libraries for environment variables, LangChain components
# 导入环境变量、LangChain组件等必要的库
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.callbacks import StdOutCallbackHandler  # Replaces verbose=True / 替代verbose=True的回调处理器


# ============================================================
# Block 2: Load Environment Variables
# 模块2：加载环境变量
# ============================================================
# Force load API keys from .env file (override existing environment variables)
# 强制从.env文件加载API密钥（覆盖已有环境变量）
load_dotenv(override=True)


# ============================================================
# Block 3: Define Prompt Template
# 模块3：定义提示词模板
# ============================================================
# Create a chat prompt template with system and human messages
# Format: system defines the role, human provides the actual query
# 创建包含系统和人类消息的聊天提示词模板
# 格式：系统消息定义角色，人类消息提供实际查询
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a {subject} teacher"),
    ("human", "Tell me about {concept}")
])


# ============================================================
# Block 4: Initialize Model and Parser
# 模块4：初始化模型和解析器
# ============================================================
# Initialize the ChatOpenAI model with specific settings
# 初始化ChatOpenAI模型及其配置
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Initialize output parser to extract plain text from complex objects
# 初始化输出解析器，从复杂对象中提取纯文本
parser = StrOutputParser()


# ============================================================
# Block 5: Build LCEL Chain (LangChain Expression Language)
# 模块5：构建LCEL链（LangChain表达式语言）
# ============================================================
# Modern approach: Use pipe operator (|) to chain components
# This replaces the deprecated LLMChain pattern
# 现代化方法：使用管道操作符（|）链接组件
# 这取代了已弃用的LLMChain模式
chain = prompt | llm | parser


# ============================================================
# Block 6: Execute with Callback Handler
# 模块6：使用回调处理器执行
# ============================================================
print("=" * 70)
print("Enabling detailed log monitoring (modern alternative to verbose=True)")
print("启用详细日志监控（verbose=True的现代替代方案）")
print("=" * 70)
print()

# Instantiate the callback handler for monitoring chain execution
# 实例化回调处理器以监控链执行过程
handler = StdOutCallbackHandler()

# Invoke the chain with input parameters and inject callback via config
# This provides X-ray vision into how data flows through the chain
# 使用输入参数调用链，并通过config注入回调
# 这提供了数据如何流经链条的透视视图
response = chain.invoke(
    {"subject": "physics", "concept": "galaxy"},
    config={'callbacks': [handler]}  # Core magic: inject callback monitoring / 核心魔法：注入回调监控
)

print()
print("=" * 70)
print("Final Output (Clean Text)")
print("最终输出（纯文本）")
print("=" * 70)
print(response)
```
