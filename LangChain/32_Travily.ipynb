```python
import os
# 【硬性要求】加载环境变量 (确保你的 .env 文件中包含 TAVILY_API_KEY=tvly-xxxx)
from dotenv import load_dotenv
load_dotenv(override=True)

# 导入 Tavily 搜索结果工具
from langchain_community.tools.tavily_search import TavilySearchResults

# ==========================================
# 1. 实例化搜索引擎工具 (Tool Initialization)
# ==========================================
# 【科研级保姆注释】
# 旧做法：手动读取 os.getenv("TAVILY_API_KEY")
# 现代做法：只要 .env 里配好了，直接实例化即可，Pydantic 会自动从环境中读取密钥。
# 默认情况下，它会返回 5 个最相关的结果。
tool = TavilySearchResults()

# ==========================================
# 2. 执行实时在线搜索 (Real-time Online Search)
# ==========================================
if __name__ == "__main__":
    # 我们用老师演示的查询：询问关于 2024 年 T20 板球世界杯的实时信息
    query = "Who is the president of the US now?"
    print(f"=== 🔍 Searching the internet via Tavily: '{query}' ===\n")
    
    # 使用现代的 .invoke() 方法调用工具
    response = tool.invoke({"query": query})
    
    # ==========================================
    # 3. 解析并验证数据结构 (Data Structure Parsing)
    # ==========================================
    print(f"✅ Search completed, a total of {len(response)} results were returned.\n")
    
    # 让我们像老师那样，探查第一条结果的具体内容
    if len(response) > 0:
        first_result = response[0]
        
        print("--- 🥇 Top-ranked search result ---")
        print(f"🔗 Source link (URL): {first_result.get('url')}")
        print(f"📄 Extracted content (Content):\n{first_result.get('content')}")
```
