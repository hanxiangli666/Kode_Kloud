```python
import os
# ã€ç¡¬æ€§è¦æ±‚ã€‘åŠ è½½ç¯å¢ƒå˜é‡ (ç¡®ä¿ä½ çš„ .env æ–‡ä»¶ä¸­åŒ…å« TAVILY_API_KEY=tvly-xxxx)
from dotenv import load_dotenv
load_dotenv(override=True)

# å¯¼å…¥ Tavily æœç´¢ç»“æœå·¥å…·
from langchain_community.tools.tavily_search import TavilySearchResults

# ==========================================
# 1. å®ä¾‹åŒ–æœç´¢å¼•æ“å·¥å…· (Tool Initialization)
# ==========================================
# ã€ç§‘ç ”çº§ä¿å§†æ³¨é‡Šã€‘
# æ—§åšæ³•ï¼šæ‰‹åŠ¨è¯»å– os.getenv("TAVILY_API_KEY")
# ç°ä»£åšæ³•ï¼šåªè¦ .env é‡Œé…å¥½äº†ï¼Œç›´æ¥å®ä¾‹åŒ–å³å¯ï¼ŒPydantic ä¼šè‡ªåŠ¨ä»ç¯å¢ƒä¸­è¯»å–å¯†é’¥ã€‚
# é»˜è®¤æƒ…å†µä¸‹ï¼Œå®ƒä¼šè¿”å› 5 ä¸ªæœ€ç›¸å…³çš„ç»“æœã€‚
tool = TavilySearchResults()

# ==========================================
# 2. æ‰§è¡Œå®æ—¶åœ¨çº¿æœç´¢ (Real-time Online Search)
# ==========================================
if __name__ == "__main__":
    # æˆ‘ä»¬ç”¨è€å¸ˆæ¼”ç¤ºçš„æŸ¥è¯¢ï¼šè¯¢é—®å…³äº 2024 å¹´ T20 æ¿çƒä¸–ç•Œæ¯çš„å®æ—¶ä¿¡æ¯
    query = "Who is the president of China now?"
    print(f"=== ğŸ” æ­£åœ¨é€šè¿‡ Tavily æœç´¢äº’è”ç½‘: '{query}' ===\n")
    
    # ä½¿ç”¨ç°ä»£çš„ .invoke() æ–¹æ³•è°ƒç”¨å·¥å…·
    response = tool.invoke({"query": query})
    
    # ==========================================
    # 3. è§£æå¹¶éªŒè¯æ•°æ®ç»“æ„ (Data Structure Parsing)
    # ==========================================
    print(f"âœ… æœç´¢å®Œæˆï¼Œå…±è¿”å›äº† {len(response)} æ¡ç»“æœã€‚\n")
    
    # è®©æˆ‘ä»¬åƒè€å¸ˆé‚£æ ·ï¼Œæ¢æŸ¥ç¬¬ä¸€æ¡ç»“æœçš„å…·ä½“å†…å®¹
    if len(response) > 0:
        first_result = response[0]
        
        print("--- ğŸ¥‡ æ’åç¬¬ä¸€çš„æœç´¢ç»“æœ ---")
        print(f"ğŸ”— æ¥æºé“¾æ¥ (URL): {first_result.get('url')}")
        print(f"ğŸ“„ æå–å†…å®¹ (Content):\n{first_result.get('content')}")
```
