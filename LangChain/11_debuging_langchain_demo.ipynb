```python
import os
from dotenv import load_dotenv

# 1. ç°ä»£åŒ–å¯¼åŒ…ï¼šæŠ›å¼ƒæ—§çš„ chainsï¼Œå¼•å…¥è§£æå™¨
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser # ğŸŒŸ æ–°å¢ï¼šä¸“é—¨ç”¨æ¥æå–å¹²å‡€æ–‡æœ¬çš„åˆ©å™¨
from langchain_core.globals import set_debug

# 0. å¿…å¤‡èµ·æ‰‹å¼ï¼šå¼ºåˆ¶åŠ è½½æœ¬åœ°ç§˜é’¥
load_dotenv(override=True)

# å¼€å¯è°ƒè¯•æ¨¡å¼ï¼Œä¸Šå¸è§†è§’å¼€å¯
set_debug(True)

# 1. å®šä¹‰æç¤ºè¯æ¨¡æ¿ (å®Œå…¨æ²¡å˜ï¼Œè¿™æ˜¯ç°ä»£æ ‡å‡†)
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a {subject} teacher"),
    ("human", "Tell me about {concept}")
])

# 2. åˆå§‹åŒ–å¤§è¯­è¨€æ¨¡å‹ (å»ºè®®å†™æ˜ model å’Œ temperatureï¼Œå…»æˆå¥½ä¹ æƒ¯)
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 3. åˆå§‹åŒ–å­—ç¬¦ä¸²è§£æå™¨
parser = StrOutputParser()

# =================================================================
# ğŸš€ 4. æ ¸å¿ƒæ”¹é€ ï¼šå½»åº•æ¶ˆç­ LLMChainï¼Œæ‹¥æŠ± LCEL ç®¡é“è¯­æ³•ï¼
# =================================================================
# ä»¥å‰ï¼šchain = LLMChain(llm=llm, prompt=prompt)
# ç°åœ¨ï¼šæç®€çš„ç»„è£…æµæ°´çº¿
chain = prompt | llm | parser

print("æ­£åœ¨å‘¼å«ç‰©ç†è€å¸ˆ...\n")

# 5. è°ƒç”¨å¹¶æ‰§è¡Œ
# å› ä¸ºæœ€ååŠ äº† parserï¼Œå¤§æ¨¡å‹åå‡ºæ¥çš„é‚£å †å¤æ‚å¯¹è±¡ç›´æ¥è¢«æ´—æˆäº†çº¯å­—ç¬¦ä¸²ï¼
response = chain.invoke({"subject": "physics", "concept": "galaxy"})

print("\n========== æœ€ç»ˆè§£ç­” ==========\n")
# ä»¥å‰ä½ éœ€è¦ç—›è‹¦åœ°æå–å­—å…¸ï¼šresponse["text"]
# ç°åœ¨ç›´æ¥æ‰“å°å°±æ˜¯æœ€å¹²å‡€çš„æ–‡æœ¬ï¼
print(response)

# æŸ¥å®Œé”™è®°å¾—æŠŠä¸Šå¸è§†è§’å…³æ‰ï¼Œä¿æŒæ§åˆ¶å°æ¸…çˆ½
set_debug(False)
```
