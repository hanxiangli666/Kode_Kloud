```python

# ============================================================================
# [部分 1] 环境初始化与依赖导入 / Part 1: Environment Setup and Imports
# ============================================================================
import os
from dotenv import load_dotenv

# ChatOpenAI：聊天模型接口
# ChatOpenAI: chat model interface
from langchain_openai import ChatOpenAI
# ChatPromptTemplate：对话提示词模板
# ChatPromptTemplate: chat prompt template
from langchain_core.prompts import ChatPromptTemplate
# StrOutputParser：把模型响应对象提取为纯字符串
# StrOutputParser: extract plain string from model response object
from langchain_core.output_parsers import StrOutputParser
# set_debug：开启/关闭 LangChain 全局调试日志
# set_debug: toggle LangChain global debug logging
from langchain_core.globals import set_debug

# 强制加载本地 .env 中的 API 密钥
# Force load API key from local .env file
load_dotenv(override=True)


# ============================================================================
# [部分 2] 开启调试模式 / Part 2: Enable Debug Mode
# ============================================================================

# 开启“上帝视角”调试：控制台会显示链路内部步骤
# Enable debug mode: console shows internal chain execution details
set_debug(True)


# ============================================================================
# [部分 3] 定义提示词模板 / Part 3: Define Prompt Template
# ============================================================================

# 使用 system + human 的标准对话模板
# Use standard chat template with system + human roles
# - {subject}: 动态角色学科
# - {concept}: 动态提问概念
#
# - {subject}: dynamic teacher subject
# - {concept}: dynamic concept to explain
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a {subject} teacher"),
    ("human", "Tell me about {concept}")
])


# ============================================================================
# [部分 4] 初始化模型与输出解析器 / Part 4: Initialize Model and Output Parser
# ============================================================================

# temperature=0：减少随机性，便于稳定演示与调试观察
# temperature=0: reduce randomness for stable demo/debug behavior
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# 字符串解析器：将 AIMessage 等对象“洗”成纯文本
# String parser: transform AIMessage-like output into plain text
parser = StrOutputParser()


# ============================================================================
# [部分 5] 用 LCEL 管道组装链条 / Part 5: Build Chain with LCEL Pipe Syntax
# ============================================================================

# 现代写法：Prompt -> LLM -> Parser
# Modern LCEL style: Prompt -> LLM -> Parser
# 旧写法常见是 LLMChain；新写法更直观、组合性更强
# Legacy style used LLMChain; LCEL is more composable and readable
chain = prompt | llm | parser

print("Calling the physics teacher...\n")


# ============================================================================
# [部分 6] 执行链路 / Part 6: Invoke the Chain
# ============================================================================

# 传入动态变量，模板会自动填充 {subject}/{concept}
# Pass runtime variables; template auto-fills {subject}/{concept}
response = chain.invoke({"subject": "physics", "concept": "galaxy"})


# ============================================================================
# [部分 7] 输出最终结果 / Part 7: Print Final Answer
# ============================================================================
print("\n========== Final Answer ==========\n")
# 因为末端接了 StrOutputParser，这里 response 已是纯字符串
# Since StrOutputParser is at the end, response is already plain text
print(response)


# ============================================================================
# [部分 8] 关闭调试模式 / Part 8: Disable Debug Mode
# ============================================================================

# 调试完成后关闭，避免影响后续 notebook 输出整洁度
# Turn debug off after inspection to keep later outputs clean
set_debug(False)
```
