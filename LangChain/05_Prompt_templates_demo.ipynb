```python

# ============================================================================
# [部分 1] 环境初始化与核心依赖导入 / Part 1: Environment Setup and Imports
# ============================================================================
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
# ChatPromptTemplate：LangChain 的核心模板引擎
# 用来管理包含占位符的提示词，实现"一个模板多种用途"
# ChatPromptTemplate: LangChain's template engine for managing reusable prompts
from langchain_core.prompts import ChatPromptTemplate

# 强制加载 .env 文件中的 API 密钥
# Force load API key from .env file
load_dotenv(override=True)


# ============================================================================
# [部分 2] 定义模板字符串（包含占位符）/ Part 2: Define Template Strings with Placeholders
# ============================================================================

# 系统消息模板：定义 AI 的角色
# System message template: defines AI's role
# {subject} 是占位符，运行时会被替换成真实的科目（如"math"、"physics"）
# {subject} is a placeholder that will be replaced at runtime (e.g., "math", "physics")
sysmsg = "You are a {subject} teacher"

# 用户消息模板：用户的实际问题
# User message template: the actual user question
# {concept} 是占位符，运行时会被替换成真实的概念（如"calculus"、"algebra"）
# {concept} is a placeholder that will be replaced at runtime (e.g., "calculus", "algebra")
humanmsg = "Tell me about {concept}"


# ============================================================================
# [部分 3] 创建 ChatPromptTemplate 对象 / Part 3: Create ChatPromptTemplate
# ============================================================================

# ChatPromptTemplate.from_messages() 是构建模板的标准方式
# It takes a list of tuples: (role, message_template)
# 这个方法会：
#   1. 识别模板中的所有占位符（{subject}, {concept}）
#   2. 创建一个 ChatPromptTemplate 对象
#   3. 保留模板的原始结构
#
# from_messages() is the standard way to build a template
# It identifies all placeholders and preserves the structure
prompt_template = ChatPromptTemplate.from_messages([
    ("system", sysmsg),      # 角色，消息内容 / Role, Message content
    ("human", humanmsg)      # 角色，消息内容 / Role, Message content
])


# ============================================================================
# [部分 4] 使用 format_prompt 填充占位符 / Part 4: Format Template with Variables
# ============================================================================

# format_prompt() 方法：用真实变量替换模板中的占位符
# format_prompt() method: replace placeholders with actual values
# 参数：subject="math"，concept="calculus"
# 这样会生成：
#   system: "You are a math teacher"
#   human: "Tell me about calculus"
#
# Arguments: subject="math", concept="calculus"
# Result:
#   system: "You are a math teacher"
#   human: "Tell me about calculus"
prompt = prompt_template.format_prompt(subject="math", concept="calculus")


# ============================================================================
# [部分 5] 检查生成的消息结构 / Part 5: Inspect Generated Message Structure
# ============================================================================

# .to_messages() 方法：查看格式化后的消息对象列表
# to_messages() method: view the formatted message objects
# 这对调试非常有用，可以看到最终会发送给 LLM 的确切内容
# Useful for debugging to see exactly what will be sent to the LLM
print("The assembled underlying message looks like this:\n", prompt.to_messages(), "\n")


# ============================================================================
# [部分 6] 初始化大语言模型 / Part 6: Initialize Language Model
# ============================================================================

# 创建 ChatOpenAI 实例
# 这时还没有 temperature 参数，会使用默认值（通常 0.7）
# Create ChatOpenAI instance
# No temperature parameter here, so it uses the default value (usually 0.7)
model = ChatOpenAI(model="gpt-4o-mini")


# ============================================================================
# [部分 7] 调用模型并获取响应 / Part 7: Invoke Model with Formatted Prompt
# ============================================================================

print("Calling the math teacher...\n")

# 发送格式化后的提示词给 LLM
# 重要：我们发送的是 prompt 对象（已格式化），而不是 prompt_template（模板）
# Send the formatted prompt (not the raw template) to the LLM
# The LLM receives:
#   System: "You are a math teacher"
#   Human: "Tell me about calculus"
response = model.invoke(prompt)


# ============================================================================
# [部分 8] 打印最终结果 / Part 8: Display Final Response
# ============================================================================

# response.content：获取 LLM 生成的文本内容
# response.content: extract the actual text from the LLM response
print(response.content)
```
