```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
# 1. 修复导包：使用最新的核心库路径
from langchain_core.prompts import ChatPromptTemplate

# 2. 必备起手式：强制加载你本地的 API 密钥
load_dotenv(override=True)

# 3. 定义包含“占位符”的字符串模板
sysmsg = "You are a {subject} teacher"
humanmsg = "Tell me about {concept}"

# 4. 修复语法错误：注意看这里的方括号和圆括号的嵌套！
prompt_template = ChatPromptTemplate.from_messages([
    ("system", sysmsg), 
    ("human", humanmsg)
])

# 5. 注入灵魂：用 format_prompt 把变量填进去
prompt = prompt_template.format_prompt(subject="math", concept="calculus")
print("组装好的底层消息长这样：\n", prompt.to_messages(), "\n")

# 6. 唤醒大模型并发送
model = ChatOpenAI(model="gpt-4o-mini")
print("正在呼叫数学老师...\n")
response = model.invoke(prompt)

# 7. 打印最终解答
print(response.content)
```
