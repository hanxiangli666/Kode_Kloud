```python
# ============================================================
# Block 1: Import Dependencies and Load Environment
# 模块1：导入依赖库并加载环境变量
# ============================================================
# Import environment management libraries
# 导入环境管理库
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from .env file
# override=True ensures the latest .env configuration is read
# 从.env文件加载环境变量
# override=True 确保读取最新的.env文件配置
load_dotenv(find_dotenv(), override=True)


# ============================================================
# Block 2: Import LangChain Components
# 模块2：导入LangChain组件
# ============================================================
# Import standardized LangChain modules
# 导入标准化的LangChain模块
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


# ============================================================
# Block 3: Define Chain Components
# 模块3：定义链组件
# ============================================================
# Create a simple prompt template for Q&A
# 创建简单的问答提示词模板
prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant.
Answer the following question: {question}
""")

# Initialize the ChatOpenAI model with temperature setting
# 初始化ChatOpenAI模型及温度设置
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

# Initialize output parser to convert responses to plain strings
# 初始化输出解析器，将响应转换为纯字符串
output_parser = StrOutputParser()


# ============================================================
# Block 4: Build LCEL Chain
# 模块4：构建LCEL链
# ============================================================
# Use pipe operator to chain components together
# 使用管道符将组件链接在一起
chain = prompt | llm | output_parser


# ============================================================
# Block 5: Prepare Batch Processing Data (Core of Demo3)
# 模块5：准备批处理数据（Demo3的核心）
# ============================================================
# Create a list of multiple questions for batch processing
# The key advantage: LangChain will process these in parallel, not sequentially
# 创建用于批处理的多个问题列表
# 关键优势：LangChain会并行处理这些问题，而非顺序执行
questions = [
    {"question": "Tell me about The Forrest Gump Movie"},
    {"question": "Tell me about Avatar Movie"}
]


# ============================================================
# Block 6: Execute Batch Processing
# 模块6：执行批处理
# ============================================================
print("=" * 70)
print("Starting Batch Processing (Parallel Execution)")
print("=" * 70)
print()

# Call batch() method to process all questions simultaneously
# This is much faster than calling invoke() multiple times
# 调用batch()方法同时处理所有问题
# 这比多次调用invoke()快得多
responses = chain.batch(questions)


# ============================================================
# Block 7: Display Results
# 模块7：显示结果
# ============================================================
# Iterate through responses (now a list of strings)
# 遍历响应结果（现在是字符串列表）
for i, res in enumerate(responses):
    print(f"Answer {i+1} / 回答 {i+1}:")
    print("-" * 70)
    print(res)
    print()

print("=" * 70)
print("Batch Processing Completed Successfully")
print("=" * 70)
```
