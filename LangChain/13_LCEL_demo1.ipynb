```python
# ============================================================
# Block 1: Import Dependencies and Load Environment
# 模块1：导入依赖库并加载环境变量
# ============================================================
# Import required libraries for environment management
# 导入环境管理所需的库
import os
from dotenv import load_dotenv, find_dotenv

# Load environment variables from .env file
# override=True ensures the latest .env configuration is read
# 从.env文件加载环境变量
# override=True 确保读取最新的.env文件配置
load_dotenv(find_dotenv(), override=True)


# ============================================================
# Block 2: Import LangChain Components
# 模块2：导入LangChain组件
# ============================================================
# Import standardized LangChain modules (official recommended packages)
# 导入标准化的LangChain模块（官方推荐的包）
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


# ============================================================
# Block 3: Define Prompt Template
# 模块3：定义提示词模板
# ============================================================
# Create a prompt template with {question} as the input variable
# The template structure: system context + user input placeholder
# 创建包含{question}作为输入变量的提示词模板
# 模板结构：系统上下文 + 用户输入占位符
prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant.
Answer the following question: {question}
""")


# ============================================================
# Block 4: Initialize Language Model
# 模块4：初始化语言模型
# ============================================================
# Initialize ChatOpenAI with specific model and temperature settings
# temperature=0.7 provides a balance between creativity and consistency
# 初始化ChatOpenAI，指定模型和温度设置
# temperature=0.7 在创造性和一致性之间提供平衡
llm = ChatOpenAI(
    model="gpt-4o-mini", 
    temperature=0.7
)


# ============================================================
# Block 5: Initialize Output Parser
# 模块5：初始化输出解析器
# ============================================================
# Create a parser to convert complex AIMessage objects to plain text strings
# 创建解析器，将复杂的AIMessage对象转换为纯文本字符串
output_parser = StrOutputParser()


# ============================================================
# Block 6: Build LCEL Chain
# 模块6：构建LCEL链
# ============================================================
# Use pipe operator "|" to chain components in sequence:
# Input -> Format Prompt -> Call LLM -> Parse Output
# 使用管道符"|"按顺序链接组件：
# 输入 -> 格式化提示词 -> 调用LLM -> 解析输出
chain = prompt | llm | output_parser


# ============================================================
# Block 7: Execute Chain and Display Results
# 模块7：执行链并显示结果
# ============================================================
# Invoke the chain with a dictionary input
# The key must match the placeholder in the template ({question})
# 使用字典输入调用链
# 键必须与模板中的占位符匹配（{question}）
response = chain.invoke({"question": "Tell me about The Forrest Gump movie."})

# Display the final response
# 显示最终响应
print("=" * 70)
print("AI Response / AI响应:")
print("=" * 70)
print(response)
print()


# ============================================================
# Block 8: Inspect Chain Schema (Optional Debugging)
# 模块8：检查链模式（可选调试）
# ============================================================
# Display input/output schema for debugging purposes
# These show the expected data structure for chain I/O
# 显示输入/输出模式用于调试
# 这些显示了链I/O的预期数据结构
print("=" * 70)
print("Input Schema / 输入模式:")
print("=" * 70)
print(chain.input_schema.model_json_schema())
print()

print("=" * 70)
print("Output Schema / 输出模式:")
print("=" * 70)
print(chain.output_schema.model_json_schema())
```
