```python
import os
from dotenv import load_dotenv, find_dotenv

# 1. 初始化环境
# 按照你的要求，使用 override=True 加载 .env
load_dotenv(find_dotenv(), override=True)

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 2. 核心链配置
prompt = ChatPromptTemplate.from_template("Answer the following question: {question}")
llm = ChatOpenAI(model="gpt-4o-mini")
output_parser = StrOutputParser()

chain = prompt | llm | output_parser

# 3. 老师强调的 Chunk 功能演示
question_content = "Tell me about The Forrest Gump movie"

print("--- 正在通过 Chunk 实时流式输出 ---")

# chain.stream 会返回一个迭代器，里面的每一个元素就是一个 chunk (分块)
# 这就是老师在视频里演示的“打字机效果”的核心
for chunk in chain.stream({"question": question_content}):
    # 这里的 chunk 是模型刚生成的最新一小段文字
    # 我们使用 print(chunk) 将其立即显示出来
    # end="" 表示打印完不换行，flush=True 确保数据不留在缓冲区直接上屏
    print(chunk, end="", flush=True)

print("\n--- 输出完毕 ---")
```
