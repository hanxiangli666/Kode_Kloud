```python

# ============================================================================
# [部分 1] 环境初始化与依赖导入 / Part 1: Environment Setup and Imports
# ============================================================================
import os
from dotenv import load_dotenv, find_dotenv

# 按要求加载 .env（override=True 表示用 .env 覆盖同名环境变量）
# Load .env as requested (override=True means .env values overwrite existing env vars)
load_dotenv(find_dotenv(), override=True)

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


# ============================================================================
# [部分 2] 组装 LCEL 核心链路 / Part 2: Build Core LCEL Pipeline
# ============================================================================

# Prompt：定义输入模板（question 是动态占位符）
# Prompt: define input template (question is a dynamic placeholder)
prompt = ChatPromptTemplate.from_template("Answer the following question: {question}")

# LLM：初始化聊天模型（此处用 gpt-4o-mini）
# LLM: initialize chat model (using gpt-4o-mini here)
llm = ChatOpenAI(model="gpt-4o-mini")

# Output Parser：将模型返回对象提取为纯文本字符串
# Output Parser: convert model response object into plain text string
output_parser = StrOutputParser()

# LCEL 管道语法：Prompt -> LLM -> Text
# LCEL pipe syntax: Prompt -> LLM -> Text
chain = prompt | llm | output_parser


# ============================================================================
# [部分 3] 准备输入问题 / Part 3: Prepare Runtime Question
# ============================================================================

# 这是运行时传给模板的真实问题内容
# This is the actual runtime question to inject into template
question_content = "Tell me about The Forrest Gump movie"


# ============================================================================
# [部分 4] 流式输出（Chunk Streaming）/ Part 4: Real-time Chunk Streaming
# ============================================================================

print("--- Outputting in real-time via Chunk streaming ---")

# chain.stream() 返回一个可迭代对象，每次产出一小段文本 chunk
# chain.stream() returns an iterator that yields small text chunks incrementally
for chunk in chain.stream({"question": question_content}):
    # end=""：不自动换行，保持打字机连续输出
    # flush=True：立即刷新缓冲区，确保实时可见
    # end="": no auto newline, keep typewriter-like continuous output
    # flush=True: flush buffer immediately for real-time rendering
    print(chunk, end="", flush=True)


# ============================================================================
# [部分 5] 收尾提示 / Part 5: Completion Marker
# ============================================================================

print("\n--- Output completed ---")
```
